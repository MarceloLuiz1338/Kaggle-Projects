***First Univariate Analysis***

```python
##Understanding the data
plt.subplot(331)
train['Survived'].value_counts(normalize=True).plot.bar(color=['r','g'])
plt.subplot(332)
train['Sex'].value_counts(normalize=True).plot.bar(figsize=(9,9),color=['b','r'])
plt.subplot(333)
train['Embarked'].value_counts(normalize=True).plot.bar(figsize=(9,9),color=['y','g','r']);
```

```python
mf=pd.DataFrame(columns=['Male','Female'])
mf.loc['Sex']=[train['Survived'][train['Sex']=='male'].mean(),train['Survived'][train['Sex']=='female'].mean()]
mf.plot.bar(figsize=(3,3), title='Surviving % per sex',rot=0);

scq=pd.DataFrame(columns=['S','C','Q'])
scq.loc['Embarked']=[train['Survived'][train['Embarked']=='S'].mean(),train['Survived'][train['Embarked']=='C'].mean(),train['Survived'][train['Embarked']=='Q'].mean()]
scq.plot.bar(figsize=(3,3), title='Surviving % per Boarding',rot=0, );
```
